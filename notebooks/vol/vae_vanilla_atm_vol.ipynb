```python
import sys
import matplotlib.pyplot as plt
import numpy as np 

sys.path.append('..')
from src.data.vol.get_atm_vols import get_atm_vols

df = get_atm_vols()
data = df.to_numpy()

data_norm = data
# data = np.expand_dims(data, -1)
# print(data.shape)

from src.models.vae_vanilla import VaeVanilla
vae = VaeVanilla(input_shape=data_norm.shape[1:],
                 hidden_layers_nodes=[100, 50, 25, 12],
                 latent_space_dim=1,
                 loss_type='mse',
                 beta=1e-2)
vae.compile(optimizer='adam')
vae.fit(data_norm, epochs=50, batch_size=100)  # 16 and 100 provide 0.0055 loss
```
